Converting the datatypes of "Join Date" and "Last Payment Date" columns from object to datetime64[ns]

## Gender analysis

```python
gen_uni=df['Gender'].unique()
no_g=[]
for i in gen_uni:
    no_g.append(df['Gender'].loc[df['Gender']==i].shape[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=gen_uni,y=no_g)
plt.title("Gender Frequency")
plt.xlabel("Gender")
plt.ylabel("Frequency")
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df['Join Date'], fill=True)
plt.title("Join Date Density")
```

As we observed in the above plot that alot of people joined Netflix subscription in 2022 according to given dataset. But from other sources we know that Netflix experienced its first subscriber loss in 2022 due to announcing plans to limit sharing. So this dataset does not have enough data for coming to this conclusion. We need more data

```python
sub_type=df['Subscription Type'].unique()
no_s=[]
for i in sub_type:
    no_s.append(df['Subscription Type'].loc[df['Subscription Type']==i].shape[0])
```

```python
sns.barplot(x=sub_type,y=no_s)
plt.xlabel("Subscription Type")
plt.ylabel("No. of Subscribers")
plt.title("How popular is each subscription plan?")
```

As we see most of the people prefer Basic plan over others. Some individuals might have the intention of testing out the service prior to making a pricier and longer-term commitment, which could account for this phenomenon. To furthur investigate the reasons for it, let's see which age group prefer which plan more.
